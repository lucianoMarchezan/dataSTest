```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from google.colab import files
import numpy as np

uri = "https://raw.githubusercontent.com/lucianoMarchezan/dataSTest/main/data/UMLToCodeICSE/cres_by_model.csv"
cres_by_model = pd.read_csv(uri)
cres_by_model 
```

```python
chart = sns.barplot(x = "size", y = "time",  hue="model_name",ci = None,data = cres_by_model, palette="ch:r=-.2,d=.3_r",
                    dodge=False,edgecolor = 'black')
plt.legend(title='Systems', loc='upper left',  fontsize = 11)
chart.set_xlabel("System's Size (number of model elements)", fontsize = 15)
chart.set_ylabel("Average CREs evaluation Time (μs)", fontsize = 15)  
fig1 = plt.gcf() 
fig1.savefig('timeBySize.pdf',dpi=300, bbox_inches='tight')
```

```python
uri = "https://raw.githubusercontent.com/lucianoMarchezan/dataSTest/main/data/UMLToCodeICSE/changes_by_model.csv"
changes_by_model = pd.read_csv(uri)  
changes_by_model['cres'] = changes_by_model['cres'].div(100).round(2) # generate average of CRES per changetype
changes_by_model.head()

crh3_filter = changes_by_model["change_type"]!='DeleteClass'
crh4_filter = changes_by_model["change_type"]!='s6'
changes_by_model_filtered = changes_by_model[(crh3_filter) & (crh4_filter)] # removing one outlier that destroys the plot
changes_by_model_filtered.head()
```

```python
uri2 = "https://raw.githubusercontent.com/lucianoMarchezan/dataSTest/main/data/UMLToCodeICSE/changes_by_model.csv" 
changes_by_model = pd.read_csv(uri2)  
changes_by_model['cres'] = changes_by_model['cres'].div(100).round(2) # generate average of CRES per changetype
changes_by_model['inconsistencies'] = changes_by_model['inconsistencies'].div(100).round(2) # generate average of CRES per changetype
chart = sns.barplot(x = "change_type", y = "cres", ci = None, data = changes_by_model,palette=["white"],dodge=False, edgecolor="black") 
chart2 = sns.barplot(x = "change_type", y = "inconsistencies", ci = None, data = changes_by_model,palette=["#6f8ca4"],dodge=False, edgecolor="black") 
plt.legend(title='Avg. Number of', loc='upper left',labels=['CREs', 'Inconsistencies'], fontsize = 12) 
chart.set_xlabel("Change Type", fontsize = 15)
chart.set_ylabel("Average", fontsize = 15)
sns.set(rc={'figure.figsize':(8,6)}) 
plt.xticks(rotation=75)
fig1 = plt.gcf() 
sns.set(rc={'figure.figsize':(8,6)}) 
fig1.savefig('changesCREs.pdf',dpi=300, bbox_inches='tight')
```
