```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from google.colab import files
import numpy as np
```

```python
uri = "https://raw.githubusercontent.com/lucianoMarchezan/dataSTest/main/data/apoSolution/systems.csv"
systems = pd.read_csv(uri, index_col=False)  
table = systems[["name" ,"size", "java_files_count", "cres", "inconsistencies"]].sort_values("inconsistencies",ascending=False)
table.to_csv("systems", index=False)
```

```python
uri = "https://raw.githubusercontent.com/lucianoMarchezan/dataSTest/main/data/apoSolution/cre_count.csv"
cre_count = pd.read_csv(uri, index_col=False)    
cre_by_crd = cre_count.groupby(['crd'])
cre_by_crd["count"].describe()[['min','max','mean','50%','std']]
#cres_by_model.to_csv("cr", index=False)
```

```python
uri = "https://raw.githubusercontent.com/lucianoMarchezan/dataSTest/main/data/apoSolution/cr_by_system.csv"
cres_by_model = pd.read_csv(uri, index_col=False)    
cres_by_model
cres_by_model.to_csv("cr", index=False)
```

```python
uri = "https://raw.githubusercontent.com/lucianoMarchezan/dataSTest/main/data/apoSolution/time_systems.csv"
tot_time_per_model = pd.read_csv(uri)
chart = sns.boxplot(x="size", y="time", data=tot_time_per_model,hue="system_name" ,palette="ch:r=-.2,d=.3_r", dodge=False,showfliers = False)
plt.legend(bbox_to_anchor=(1, 1),loc=2, title='Systems',fontsize = 10)
chart.set_xlabel("System's Size (number of model elements)", fontsize = 15)
chart.set_ylabel("CREs evaluation time (Î¼s)", fontsize = 15)  
plt.xticks(rotation=75) 
sns.set(rc={'figure.figsize':(8,6)}) 
fig1 = plt.gcf() 
fig1.savefig('timeBySize.pdf',dpi=300, bbox_inches='tight')
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
uri = "https://raw.githubusercontent.com/lucianoMarchezan/dataSTest/main/data/apoSolution/repair_results.csv"
repairs = pd.read_csv(uri, index_col=False)    
repairs_by_system=repairs.groupby(['system_name'])
repairs_by_system["abstract_ras_count"].describe()[['min','max','mean','50%','std','count']]
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
uri = "https://raw.githubusercontent.com/lucianoMarchezan/dataSTest/main/data/apoSolution/Correlation%20of%20Smells%20and%20Products%20-%20Complete.csv"
correl = pd.read_csv(uri, index_col=False)       
corre_file = correl.describe()
corre_file.to_csv("corre_file",float_format='%.1f')
```

```python
uri = "https://raw.githubusercontent.com/lucianoMarchezan/dataSTest/main/data/apoSolution/repair_time_size.csv"
tot_time_per_model = pd.read_csv(uri)
chart = sns.boxplot(x="size", y="repair_generation_time", data=tot_time_per_model,hue="system_name" ,palette="ch:r=-.2,d=.3_r", dodge=False,showfliers = False)
plt.legend(bbox_to_anchor=(1, 1),loc=2, title='Systems',fontsize = 10)
chart.set_xlabel("System's Size (number of model elements)", fontsize = 15)
chart.set_ylabel("Repair generation time (Î¼s)", fontsize = 15)  
plt.xticks(rotation=75) 
sns.set(rc={'figure.figsize':(8,6)}) 
fig1 = plt.gcf() 
fig1.savefig('repairTimeBySize.pdf',dpi=300, bbox_inches='tight')
```
